```python
import pandas as pd
```

```python
scans_per_month = pd.read_csv("https://raw.githubusercontent.com/scanninglabor/IAScanningLabor/main/code/scans_per_center_per_year.csv")
```

```python
scans_per_month
```

```python
grouped_df = scans_per_month.groupby(['name', 'lat', 'long'])['count'].sum().reset_index(name='count')
```

```python
i = 0
geocode = []
for i in range(len(grouped_df)):
    geocode.append(
    str(grouped_df.iloc[i]['lat']) + ',' + str(grouped_df.iloc[i]['long']))
```

```python
grouped_df['geocode'] = geocode
```
