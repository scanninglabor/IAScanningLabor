```python
ia_texts = pd.read_csv("/Volumes/Samsung_T5/scanning_labor_in_IA/texts-data.csv")
```

```python
len(ia_texts['identifier'].unique())
ia_texts.drop_duplicates()
```

```python
ia_texts = ia_texts.drop(columns=['Unnamed: 0.2', 'Unnamed: 0.1', 'Unnamed: 0'])

ia_texts
```

```python

ia_texts['month_year'] = pd.to_datetime(ia_texts['scandate']).dt.to_period('M')
```

```python
ia_texts
```

```python
ia_scans_per_month = ia_texts.groupby(['scanningcenter','month_year']).count().reset_index()
```

```python
ia_scans_per_month
```

```python
scans_per_center = ia_scans_per_month[['scanningcenter', 'month_year', 'identifier']]
```

```python
scans_per_center = scans_per_center.rename(columns={"identifier": "scans_per_month"})
```
